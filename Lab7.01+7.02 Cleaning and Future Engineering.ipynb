There are quite a number of values in the DMA, MSA, and ADI columns and therefore we're doing a further analysis on whethere there is any correlation between these columns 

```python
data_corr = data[['DMA','MSA','ADI','TARGET_D']]
data_corr.head()
```

```python
correlations_matrix = data_corr.corr() 
correlations_matrix
```

```python
data_corr.describe().T
```

We can see that the mean is quite similar and close to the median, therefore have decided to change the NaN values to Mean for all 3 columns

```python
cols = ['DMA','MSA','ADI']
for item in cols:
    numerical[item] = numerical[item].fillna(np.ceil(np.mean(numerical[item])))
numerical[item]
```

Seeing that there are no null values!

```python
for item in cols:
    print(numerical[item].isna().sum())
```
